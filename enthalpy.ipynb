```python
import pandas as pd 
import matplotlib.pyplot as plt

df_location=pd.read_csv('output/location.txt',delim_whitespace=True)
gamma_cycle=[]
timestep=[]

########
 ########
plt.figure(figsize=(8,8))
for k in range(4):
    ###################################
    ###################################
    df=pd.read_csv('output/log_lammps_recognition%d.txt'%k, delim_whitespace=True)
    LOG0=pd.DataFrame(())
    for index, i in df.iterrows(): 
        print(i['finish'])
        LOG1=pd.read_csv('data/cycles/%dcycles/log.lammps'%k, delim_whitespace=True, index_col=False, skiprows=int(i['start'])+1, nrows=(int(i['finish'])-int(i['start']))-7, names=['Step', 'TotEng', 'PotEng', 'KinEng', 'Temp' ,'Press' ,'E_pair' ,'E_vdwl', 'E_coul' ,'E_long' ,'E_bond', 'Fnorm', 'Fmax','Volume' ,'Enthalpy' ,'CellAlpha' ,'CellBeta' ,'CellGamma' ,'Cella' 'Cellb' ,'Cellc '], decimal='.')
        LOG0=pd.concat([LOG0,LOG1], ignore_index=True)
    LOG0 = LOG0.apply(pd.to_numeric)
    display(LOG0)
    enthalpy=LOG0['Enthalpy'].tolist()
    step=LOG0['Step'].tolist()
    # Program to calculate moving average
    window_size = 100
    
    i = 0
    # Initialize an empty list to store moving averages
    moving_averages = []
    timestep=[]
    # Loop through the array to consider
    # every window of size 3
    while i < len(enthalpy) - window_size + 1:

        # Store elements from i to i+window_size
        # in list to get the current window
        window = enthalpy[i : i + window_size]
        timestep+=[step[i]*0.000002]
        # Calculate the average of current window
        window_average = round(sum(window) / window_size, 2)
        # Store the average of current
        # window in moving average list
        moving_averages.append(window_average)
        # Shift window to right by one position
        i+= 1
    plt.scatter(timestep, moving_averages, s=5, alpha=0.2 )
    plt.plot(timestep, moving_averages,alpha=1, label='%d Cycles'%k)
plt.title('Enthalpy of the structure against time - Moving average', fontsize=16)
plt.xlabel(r'Time    [$ns$]', fontsize=14)
plt.ylabel(r'Enthalpy   [$eV$]', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend()
plt.show()
```
