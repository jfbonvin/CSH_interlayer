```python
import pandas as pd

def locate_neighbors(N,M,L):
    df_location=pd.DataFrame((), columns=['a0','a1','b0','b1'])
    a0=[]
    a1=[]
    b0=[]
    b1=[]
    a1b=[]
    a0b=[]
    b1b=[]
    b0b=[]
    for i in range(0,N*M*L):
        ##### b0-direction check
        if i%(M*L)==0 or (i-1)%(M*L)==0: 
            b0+=['-']
            b0b+=[i+(N-1)*L]
            if (i+1)%(M*L)==0 or (i+2)%(M*L)==0: ### Look +1 and +2 because cells are enumerated from 0 to M*N*L-1
                b1+=['-']
                b1b+=[i-(N-1)*L]
            else:
                b1+=[i+L]
                b1b+=['-']
        else :
            b0+=[i-L]
            b0b+=['-']
            if (i+1)%(M*L)==0 or (i+2)%(M*L)==0:
                b1+=['-']
                b1b+=[i-(N-1)*L]
            else:
                b1+=[i+L]
                b1b+=['-']
                
        if i<M*L:
            a1+=[i+(M*L)]
            a1b+=['-']
            a0+=['-']
            a0b+=[i+(M*L)*(N-1)]
        if i>=M*L:
            a0+=[i-(M*L)]
            a0b+=['-']
            if i>((M*L)*(N-1)-1):
                a1+=['-']
                a1b+=[i-(M-1)*N*L]
            else:
                a1+=[i+(M*L)]
                a1b+=['-']
    df_location['a0b']=a0b
    df_location['a1b']=a1b
    df_location['b1b']=b1b
    df_location['b0b']=b0b
    df_location['a0']=a0
    df_location['a1']=a1
    df_location['b0']=b0
    df_location['b1']=b1
    return df_location
```

```python
df=locate_neighbors(4,4,2)
display(df)
```
